###### 2.Making a Word Cloud

```python
import nltk
from nltk.corpus import webtext
from nltk.probability import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt
 
nltk.download('webtext')
words = webtext.words('C://cloudNLP.txt')  # data in text file
data= nltk.FreqDist(words)
 
filter_words = dict([(m, n) for m, n in data.items() if len(m) > 3])
 
cloud = WordCloud().generate_from_frequencies(filter_words)
 
plt.imshow(cloud, interpolation="bilinear")
 
plt.axis("off")
(-0.5, 399.5, 199.5, -0.5)
plt.show()
```

##### 3.Extracting Nouns from a sentence

```python
nouns = TextBlob("India is a country in the Asia. This is where Apoorv lives")
 
for noun in nouns.noun_phrases:
    print(noun)
```

###### 4. Sentimental Analysis

```python
from textblob.sentiments import NaiveBayesAnalyzer
 
text = TextBlob("The movie was excellent!", analyzer=NaiveBayesAnalyzer())
print(text.sentiment)
 
text = TextBlob("The movie story pinch me a lot.", analyzer=NaiveBayesAnalyzer())
print(text.sentiment)
 
```

###### 5. Language Detection and translation

```python
blob = TextBlob("My name is Apoorv Tyagi")
 
print(blob.detect_language())
 
print(blob.translate(to='hi'))
print(blob.translate(to='fr'))
```

```python
## Translate Using 'goslate'
import goslate
 
text = "My name is Apoorv Tyagi"
 
gs = goslate.Goslate()
 
translatedText = gs.translate(text, 'fr')
print(translatedText)
 
translatedText = gs.translate(text, 'de')
print(translatedText)

translatedText = gs.translate(text, 'hi')
print(translatedText)
```

###### 6.Finding Antonym

```python
from textblob import Word
 
text_word = Word('danger')
 
antonyms = set()
for synset in text_word.synsets:
    for lemma in synset.lemmas():        
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())        
 
print(antonyms)
```

###### 7. Word Defination and Synonyms

```python
input_word='danger'
text_word = Word(input_word)
 
print(text_word.definitions)
 
synonyms = set()
for synset in text_word.synsets:
    for lemma in synset.lemmas():
        if(lemma.name()!=input_word):
            synonyms.add(lemma.name())
         
print(synonyms)
```
